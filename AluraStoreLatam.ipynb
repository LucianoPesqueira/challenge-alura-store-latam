### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 1. Análisis de facturación



```python
tiendas = {
    'tienda 1': tienda,
    'tienda 2': tienda2,
    'tienda 3': tienda3,
    'tienda 4': tienda4
}
```

```python
facturacion_total_tienda = {}
for name, df in tiendas.items():
    facturacion_total_tienda[name] = df['Precio'].sum()
    print(f"Las ventas realizadas por la {name}: ${facturacion_total_tienda[name]:,.2f}")
    
```

# 2. Ventas por categoría

```python
facturacion_ventas_categoria = {}
popularidad_ventas_categoria = {}

print("Facturacion y popularidad de las ventas por categoria")
for name, df in tiendas.items():
    facturacion_ventas_categoria[name] = df.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
    
    print(f"\nFacturacion por categoria: {name}")
    for categoria, precio in facturacion_ventas_categoria[name].items():
        print(f"- {categoria}: ${precio:,.2f}")
    
    print(f"\nCantidad por categoria: {name}")
    popularidad_ventas_categoria[name] = df['Categoría del Producto'].value_counts()
    
    for categoria, cantidad in popularidad_ventas_categoria[name].items():
        print(f"- {categoria}: {cantidad}")
    
    print("\n ---------------------------------")
```

# 3. Calificación promedio de la tienda


```python
calificaciones_promedios_tiendas = {}
print("Promedios de calificaciones por tienda")
for name, df in tiendas.items():
    calificaciones_promedios_tiendas[name] = round(df['Calificación'].mean(), 2)
    print(f"calificacion promedio de la {name} es de {calificaciones_promedios_tiendas[name]} estrellas")
```

# 4. Productos más y menos vendidos

```python
producto_mas_vendido = {}
producto_menos_vendido = {}

print("El producto mas y menos vendido por tienda")
for name, df in tiendas.items():
    print(f"{name}:")
    counts = df['Producto'].value_counts()
    producto_mas_vendido[name] = {counts.idxmax(): counts.max()}
    producto_menos_vendido[name] = {counts.idxmin(): counts.min()}
    print(f"- Producto mas vendido: {counts.idxmax()}: {counts.max()} unidades")
    print(f"- Producto menos vendido: {counts.idxmin()}: {counts.min()} unidades")
```

# 5. Envío promedio por tienda

```python
envio_promedio_tienda = {}

print("Costo del envio promedio")
for name, df in tiendas.items():
    envio_promedio_tienda[name] = round(df['Costo de envío'].mean(), 2)
    
    print(f"- Costo de envio promedio de la {name}: ${envio_promedio_tienda[name]:,.2f}")
```

# Generando gráfico

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
#c = ['teal', 'orange', 'lightcoral', 'aqua', 'gray', 'blue', 'red', 'violet'] 
c = ['#FF7700','#F0BA19','#C8B592','#D4D3C4','#69D1C5','#33658A','#2F4858','#243742']
```

### Grafico de ingresos totales por tienda

```python
#plt.bar(facturacion_total_tienda.keys(), facturacion_total_tienda.values(), color = c)
labels = list(facturacion_total_tienda.keys())
values = list(facturacion_total_tienda.values())

fig, ax = plt.subplots(1, figsize=(8, 6))
rects = ax.bar(labels, values, color=c[4])
ax.bar_label(
    rects,
    labels=[f"${v:,.0f}" for v in values],
    padding=2,
    color='black'
)

plt.title("Ingresos totales por tienda", fontweight='bold')
plt.ylabel("Ingresos (millones de ARS)")
ax.ticklabel_format(style='plain', axis='y')#desactivar notacion cientifica
plt.show()
```

### Grafico de Ventas por categoria

```python
#facturacion_ventas_categoria 
    #{'tienda 1': Categoría del Producto
    #Electrónicos               429493500.0
rows = []
for tienda, datos in facturacion_ventas_categoria.items(): #tienda: 'tienda 1' datos: Categoría del Producto Electrónicos  429493500.0
    for producto, precio in datos.items():#tipo: Electrónicos, productos: 429493500.0
        rows.append({
            'Tienda': tienda,
            'Producto': producto,
            'Precio': precio
        })
            
df_ventas = pd.DataFrame(rows)
df_ventas['Precio'] = df_ventas['Precio'].astype(float)

plt.figure(figsize=(10, 5))
sns.barplot(data= df_ventas, x='Precio', y='Producto', hue='Tienda', errorbar=None)
plt.title(f"Facturacion por categoria y tienda")
plt.xlabel("Precio (ARS)")
plt.ylabel("Categoria")
plt.gca().xaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x/1e8:,.1f}M'))
plt.show()
```

```python
df_ventas = pd.DataFrame(popularidad_ventas_categoria).T
df_ventas.plot(kind='barh', figsize=(14,8), color=c)
plt.title('Popularidad de ventas por categoria')
#plt.ylabel('Categoria del producto')
plt.xlabel('Cantidad de ventas')
plt.xticks(rotation= 45)
plt.show()
```

### Calificaciones promedio por tienda

```python
plt.bar(calificaciones_promedios_tiendas.keys(), calificaciones_promedios_tiendas.values(), color = c)
plt.title("Valoracion media por tienda ")
plt.ylabel("calificacion(promedio)")
plt.show()
```

### Graficos del Top 1 productos mas y menos vendidos por tienda

```python
rows = []
for tienda in tiendas.keys():
    for producto, cantidad in producto_mas_vendido[tienda].items():
        rows.append({
            'Tienda': tienda,
            'Producto': producto,
            'Cantidad': cantidad,
            'Tipo': 'mas vendido'
        })
    for producto, cantidad in producto_menos_vendido[tienda].items():
        rows.append({
            'Tienda': tienda,
            'Producto': producto,
            'Cantidad': cantidad,
            'Tipo': 'menos vendido'
        })
df_extremos = pd.DataFrame(rows)

plt.figure(figsize=(10, 5))
sns.barplot(data=df_extremos, x='Producto', y='Cantidad', hue='Tienda', errorbar=None)
plt.title(f"Los productos mas y menos vendidos por tienda")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Grafico de envio promedio por tienda

```python
#envio_promedio_tienda
tiendas = list(envio_promedio_tienda.keys())
valores = [float(v) for v in envio_promedio_tienda.values()]
plt.bar(tiendas, valores, color = c)
plt.title("Costo de envio promedio por tienda")
plt.show()
```
